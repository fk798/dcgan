```python
# import modules
from lshashpy3 import LSHash
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# loading data from the .npy files
def load_data(filename):
    return np.load(filename)
```

```python
images = []
for i in range(235):
    images.append(load_data("./images_{}.npy".format(i)))
print(images[0].shape)

# combine all data into one large array
imgs_np = np.concatenate(images, axis = 0)
print(imgs_np.shape)
```

```python
# lsh creation and population - extra_data is the index of the image in imgs_np
lsh = LSHash(6, 128*128*3)
for i in range(len(imgs_np)):
    lsh.index(imgs_np[i].flatten(), extra_data = i)
```

```python
# load the DCGAN generated images and query from lsh
generated = load_data("./images_generated.npy")
images_queried = lsh.query(generated[0].flatten())
```

```python
images_queried = lsh.query(generated[36].flatten())
```

```python
# number of images that are "similar" to queried image
print(len(images_queried))
```

```python
# get index of an image that is similar to queried image
ind_sim = images_queried[2][-2][-1]
```

```python
print(imgs_np[ind_sim].shape)
```

```python
# plot image from index obtained before
plt.imshow(np.transpose(imgs_np[ind_sim],(1,2,0)))
```

```python
# plot generated image just to compare their similarities
plt.imshow(np.transpose(generated[36],(1,2,0)))
```
